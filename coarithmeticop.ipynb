```python
import os
import cv2
import matplotlib.pyplot as plt


class AirthmeticOperations:
    def __init__(self, path01, path02):
        self.image_path_src = path01
        self.image_path_dest = path02

    def weight(self):
        """
            cv2.addWeighted(img1, wt1, img2, wt2, gammaValue)
            gammaValue : measurement of light
        """
        image_src = cv2.imread(self.image_path_src) ## Read the images
        image_dest = cv2.imread(self.image_path_dest)

        # resizing both of images to same dimension before adding
        # take (1050,1610) as a size we need to resize
        image_src = cv2.resize(image_src, (1050, 1610))
        image_dest = cv2.resize(image_dest, (1050, 1610))

        weightsum = cv2.addWeighted(image_src, 0.5, image_dest, 0.4, 0) ## perform blending with addWeighted function
        cv2.imwrite(os.path.join("added.jpg"),   weightsum)

    def substract(self):
        """
        substract pixel wise one image from another image
        """
        image_src = cv2.imread(self.image_path_src)
        image_dest = cv2.imread(self.image_path_dest)

        # resizing both of images to same dimension before subtracting
        # take (1050,1610) as a size we need to resize
        image_src = cv2.resize(image_src, (1050, 1610))
        image_dest = cv2.resize(image_dest, (1050, 1610))

        remaining_image = cv2.subtract(image_src, image_dest) ## Substract images with substract function
        cv2.imwrite(os.path.join("substracted.jpg"), remaining_image)
       
```

```python
# arithmetic operations on images
image_path_src = os.path.join("test.jpg")
image_path_dest = os.path.join("test02.jpg")
airthmetic_obj = AirthmeticOperations(image_path_src, image_path_dest)
airthmetic_obj.weight() 
airthmetic_obj.substract() 
```

```python
## Function plt_show to display the images 
import matplotlib.pyplot as plt
def plt_imshow(title, image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
## Read the images
image0 = cv2.imread("test.jpg")
image1 = cv2.imread("test02.jpg")
image2 = cv2.imread('added.jpg')
image3 = cv2.imread('substracted.jpg')
```

```python
## Display the images
plt_imshow("source image",image0)
plt_imshow("destination image",image1)
plt_imshow("blending image",image2)
plt_imshow("substracted image",image3)
```
