```python
import cv2
import os


class CannyEdgeDetection:
    def __init__(self, path):
        self.image_path = path

    def canny_edge(self):
        """Second and third arguments are our minVal and maxVal """
        """minval and maxval are used for Hysteresis Thresholding """

        img = cv2.imread(self.image_path, 0)
        edges = cv2.Canny(img, 50, 200)
        cv2.imwrite(os.path.join("edges.jpg"), edges)
# canny edge detection
image_path_src = os.path.join("test.jpg")
canny_edge_obj = CannyEdgeDetection(image_path_src)
canny_edge_obj.canny_edge()
```

```python
import matplotlib.pyplot as plt
def plt_imshow(image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.grid(False)
	plt.show()
```
