```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
image = cv.imread('test.jpg',0)
# ORB detector initition
orb = cv.ORB_create()
# detection of keypoints with ORB
kep = orb.detect(image,None)
# descriptors computation with ORB
kep, des = orb.compute(image, kep)
#  keypoints location,
image2 = cv.drawKeypoints(image, kep, None, color=(0,255,0), flags=0)
plt.imshow(image2), plt.show()
```
